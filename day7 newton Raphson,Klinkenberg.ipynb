```python
n = float(input("hey,enter your number to be square root: "))
x_0 = float(input("hey,enter the initial guess of root: "))

x = x_0

count = 0
print(f"the value of x_{count} is {x}")

while (abs(x**2 - n)>0.00001):
    print("====================")
    x = x-((x**2-n)/(2*x))
    count = count+1
    print(f"the value of x_{count} is {x}")

print(f"\n \n Result: The square root of {n} is {x}")
print(f"\n \n The Actual Square Root of {n} is {n**0.5}")
print(f'\n The num of iterations used = {count}')
    
```

## Klinkenberg

k(g) = k(l) + c(1/pm) 

kg = measured gas permeability pm = mean pressure kL = equivalent liquid permeability, i.e., absolute permeability, k c = slope of the line

c = b.k(l) = 6.9 kL^(-0.36)

$ 6.9(k_l)^{0.64} + p_m * k_L - p_m * k_g = 0$  = f(k_l)

f'(k_l) = 4.416*k_l^(-0.36) + p_m 

**Example : - The permeability of a core plug is measured by air. Only one measurement is made at a mean pressure of 2.152 psi. The air permeability is 46.6 md. Estimate the absolute permeability of the core sample. Compare the result with the actual absolute permeability of 23.66 md.**

```python
pm = 2.152 
kg = 46.6 

k= float(input("enter your initial guess absolute permeability kl "))

count = 0
print(f"the value of kl_{count} is {k}")

while (abs(6.9*(k**0.64)+pm*k-pm*kg)>0.00001):
    print("=======================")
    updated_value = (6.9*(k**0.64)+pm*k-pm*kg)/(4.416*k**(-0.36)+pm)
    k = k-updated_value
    count=count+1
    print(f"the value of kl_{count} is {k}")

print(f"the final value of perm k : {k}")
print(f"the absolute value of k is : {23.66}")
print(f"The % Error is : {100*abs((k-23.66)/k)}")
```
