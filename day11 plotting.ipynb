### Relative permeability plots

```python
sw = np.linspace(0,1,100)
```

```python
sw
```

```python
krw = sw**3
kro = (1-sw)**2
```

```python
krw
```

```python
kro
```

```python
# step:1 krw vs sw plot
```

```python
plt.plot(sw,krw)
```

```python
# step:2 kro vs sw
plt.plot(sw,kro)
```

```python
# step:3 krw and kro both are in same plot
plt.plot(sw,krw)
plt.plot(sw,kro)
```

```python
#step:4 give x axis label
plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("water saturation(fraction)")
```

```python
#step:5 give y axis label
plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("water saturation(fraction)")
plt.ylabel("relative permeability(fraction)")
```

```python
#step:6 making grid
plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("water saturation(fraction)")
plt.ylabel("relative permeability(fraction)")
plt.grid()
```

```python
#step:7 adding lagend
plt.plot(sw,krw, label = "water relative permeability")
plt.plot(sw,kro, label = "oil relative permeability")

plt.xlabel("water saturation(fraction)")
plt.ylabel("relative permeability(fraction)")
plt.legend()
plt.grid()
```

```python
#step:8 changing the size of plot

plt.figure(figsize=(8,6))

plt.plot(sw,krw, label = "water relative permeability")
plt.plot(sw,kro, label = "oil relative permeability")

plt.xlabel("water saturation(fraction)")
plt.ylabel("relative permeability(fraction)")
plt.legend()
plt.grid()
```

```python
#step:9 changing the colours of line
plt.figure(figsize  =(8,6))
plt.plot(sw,krw, label = "water relative permeability", color = "yellow")
plt.plot(sw,kro, label = "oil relative permeability", color = "brown")

plt.xlabel("water saturation(fraction)")
plt.ylabel("relative permeability(fraction)")
plt.legend()
plt.grid()
```

```python
#step: 10 changing the line thickness
plt.figure(figsize =(8,6))
plt.plot(sw,krw, label = "water relative permeability", color = "yellow",lw = 1)
plt.plot(sw,kro, label = "oil relative permeability", color = "brown",lw = 5)

plt.xlabel("water saturation(fraction)")
plt.ylabel("relative permeability(fraction)")
plt.legend()
plt.grid()
```

```python
#step:11 adding title to the plot
plt.figure(figsize =(8,6))
plt.plot(sw,krw, label = "water relative permeability", color = "yellow",lw = 1)
plt.plot(sw,kro, label = "oil relative permeability", color = "brown",lw = 5)

plt.xlabel("water saturation(fraction)")
plt.ylabel("relative permeability(fraction)")
plt.legend()
plt.title("relative permeability plot")
plt.grid()
```

```python
## step:12 styles
plt.style.available
```

```python
plt.style.use("grayscale")

plt.figure(figsize =(8,6))
plt.plot(sw,krw, label = "water relative permeability", color = "yellow",lw = 1)
plt.plot(sw,kro, label = "oil relative permeability", color = "brown",lw = 5)

plt.xlabel("water saturation(fraction)")
plt.ylabel("relative permeability(fraction)")
plt.legend()
plt.title("relative permeability plot")
plt.grid()
```
